# Loading datas from .ods

```python
# https://pypi.org/project/pandas-ods-reader/
# !pip install pandas_ods_reader
from pandas_ods_reader import read_ods

import pandas as pd
import numpy as np

# display non-truncated dataframe
pd.set_option('display.max_colwidth', -1)
```

```python
path = "data/88milSMS_88522.ods"

# load a sheet based on its index (1 based)
sheet_idx = 1
df = read_ods(path, sheet_idx)

# load a sheet based on its name
sheet_name = "Feuille1"
df = read_ods(path, sheet_name)

# load a file that does not contain a header row
# if no columns are provided, they will be numbered
df = read_ods(path, 1, headers=True)

# load a file and provide custom column names
# if headers is True (the default), the header row will be overwritten
# df = read_ods(path, 1, columns=["NUM_SMS", "DATETIME", "ID_NUM_TEL", "SMS_ANON"])
```

## Convert types and format date

```python
df["NUM_SMS"] = df["NUM_SMS"].astype('int64')
df["ID_NUM_TEL"] = df["ID_NUM_TEL"].astype('int64')

df['day'] = (df["DATETIME"].str.split(" ", n = 3, expand = True))[0]
df['month'] = (df["DATETIME"].str.split(" ", n = 3, expand = True))[1]
df['year'] = (df["DATETIME"].str.split(" ", n = 3, expand = True))[2]
df['time'] = (df["DATETIME"].str.split(" ", n = 3, expand = True))[3]

df["day"] = df["day"].astype('int64')
df["year"] = df["year"].astype('int64')

def month_to_num(item):
    MONTH_NUMS = {
        'jan.': 1,
        'fév.': 2,
        'mar.': 3,
        'avr.': 4,
        'mai.': 5,
        'juin.': 6,
        'juil.': 7,
        'aout.': 8,
        'sept.': 9,
        'oct.': 10,
        'nov.': 11,
        'déc.': 12,
    }
    return MONTH_NUMS.get(item)

df['month'] = df.apply(lambda x: month_to_num(x['month']),axis=1)
```

```python
df.dtypes
```

```python
# Split date
df['hour'] = (df["time"].str.split(":", n = 2, expand = True))[0]
df['minute'] = (df["time"].str.split(":", n = 2, expand = True))[1]
df['second'] = (df["time"].str.split(":", n = 2, expand = True))[2]

# Create new date
df['date_time'] = pd.to_datetime(df[['year', 'month', 'day', 'hour', 'minute', 'second']])
```

```python
df.dtypes
```

## Création timestamp

```python
df['time_stamp'] = df.date_time.values.astype(np.int64)
```

```python
df.drop(['DATETIME', 'day', 'month', 'year', 'time', 'hour', 'minute', 'second'], axis=1, inplace=True)
```

## Connect to database

```python
# https://pynative.com/python-mysql-database-connection/
# !pip install mysql-connector-python

import json
with open('config.json') as json_file:
    dbconf = json.load(json_file)
import mysql.connector
from mysql.connector import Error

# Open connection
def connect_db(dbconf):
    try:
        connection = mysql.connector.connect(host=dbconf["host"],
                                             database=dbconf["database"],
                                             user=dbconf["user"],
                                             password=dbconf["password"])
        if connection.is_connected():
            db_Info = connection.get_server_info()
            print("Connected to MySQL Server version ", db_Info)
            cursor = connection.cursor()
            cursor.execute("select database();")
            record = cursor.fetchone()
            print("You're connected to database: ", record)

    except Error as e:
        print("Error while connecting to MySQL", e)
    finally:
        if (connection.is_connected()):
            return connection
        
# Close connection
def close_connect(connection):
    if (connection.is_connected()):
        cursor.close()
        connection.close()
        print("MySQL connection is closed")
    else:
        print("MySQL error closing connection")
    
conn = connect_db(dbconf)
close_connect(conn)
```

## List columns from table

```python
table_name = "SMS"

#  List columns from table
def list_columns(table_name):

    try:
        conn = connect_db(dbconf)
        cursor = conn.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        print("You're connected to database: ", record)
        sql = "SELECT COLUMN_NAME FROM INFORMATION_SCHEMA.COLUMNS WHERE TABLE_SCHEMA = 'master_p' AND TABLE_NAME = 'SMS';"
        cursor.execute(sql)
        # Fetch all the records
        result = cursor.fetchall()
        #         print(result)
        for i in result:
            print(i)

    except Error as e:
        print("Error while connecting to MySQL", e)
    finally:
        if (connection.is_connected()):
            close_connect(conn)
   
list_columns(table_name)
```

## Insert

```python
# db_cols = "id_sms,datetime,message,UTILISATEUR_id_utilisateur,SUJET_id_sujet,ACTIVITE_id_activite,ACTIVITE_SUJET_id_sujet"
# print(toyset[:1])

def insert_toyset(toyset):

    try:    
        conn = connect_db(dbconf)
        cursor = conn.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        
        # Insert DataFrame recrds one by one.
        for i,row in toyset.iterrows():
#             print(row)
            sql = "INSERT INTO SMS(id_sms,date_time,content,time_stamp,UTILISATEUR_id_utilisateur) VALUES (%s,%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))

            # the connection is not autocommitted by default, so we must commit to save our changes
            conn.commit()
                
    except Error as e:
        print("Error while inserting to MySQL", e)
    finally:
        if (conn.is_connected()):
            cursor.close()
            # close connection
            close_connect(conn)

insert_toyset(toyset)
```

## Select from database

```python
def select_sms():
    
    try:
        conn = connect_db(dbconf)
        cursor = conn.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
                
        # Select 
        sql = "SELECT * FROM SMS;"
        cursor.execute(sql)
        
        print("\n")
        myresult = cursor.fetchall()
        for x in myresult:
          print(x, "\n")
        
        # close connection
        close_connect(conn)

    except Error as e:
        print("Error while selecting to MySQL", e)

    finally:
        if (conn.is_connected()):
            cursor.close()
            conn.close()
            
select_sms()
```
