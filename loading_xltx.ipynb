```python
import pandas as pd

# display non-truncated dataframe
pd.set_option('display.max_colwidth', -1)
```

```python
path = "data/88milSMS_88522.xltx"
df_raw = pd.read_excel(path, sheet_name="Feuille1", dtype={'DATETIME': str,'SMS_ANON': str})
```

```python
df = df_raw.drop(columns=['Unnamed: 3'])
```

```python
# df['date'] = pd.to_datetime(df['DATETIME'])
# df['date'] = df['DATETIME'].astype('str')
df['DATETIME'] = df['DATETIME'].apply(str)
```

```python
df['DATETIME'].sample(5)
```

```python
df["NUM_SMS"] = df["NUM_SMS"].astype('int64')
df["ID_NUM_TEL"] = df["ID_NUM_TEL"].astype('int64')
```

```python
df['date']  = df["DATETIME"].apply(lambda x: str(x))
```

```python
df.dtypes
```

```python
print(df['date'] .sample() )
```

```python
df['day'] = (df["DATETIME"].str.split(" ", n = 3, expand = True))[0]
df['month'] = (df["DATETIME"].str.split(" ", n = 3, expand = True))[1]
df['year'] = (df["DATETIME"].str.split(" ", n = 3, expand = True))[2]
df['time'] = (df["DATETIME"].str.split(" ", n = 3, expand = True))[3]
```
